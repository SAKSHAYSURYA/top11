```python
columns_to_drop = ["date","stage","venue","fielder","matchno","innings"]
df1 = df.drop(columns=columns_to_drop,errors='ignore')
df1
```

```python
batting_df = df1[['striker', 'batting_team']].rename(columns={
    'striker': 'player',
    'batting_team': 'team'
})

bowler_df = df1[['bowler', 'bowling_team']].rename(columns={
    'bowler': 'player',
    'bowling_team': 'team'
})
```

```python
df2 = pd.concat([batting_df, bowler_df], ignore_index=True)
df2
```

```python
df2.drop_duplicates(inplace=True)
df2
```

```python
df3 = pd.DataFrame(columns=["player","team","run_rate","bowling_avg","category"])
df3
```

```python
df3[["player","team"]]=df2[["player","team"]]
df3
```

```python
# Standardize names for accurate merge
df3['player'] = df3['player'].str.strip().str.lower()
df['striker'] = df['striker'].str.strip().str.lower()

# Calculate total runs and balls faced by each striker
runs = df.groupby('striker')['runs_of_bat'].sum()
balls = df.groupby('striker').size()
strike_rate = (runs / balls) * 100

# Prepare run rate DataFrame
strike_rate_df = strike_rate.reset_index().rename(columns={'striker': 'player', 0: 'run_rate'})

# Merge into df3 without creating new columns like run_rate_x or run_rate_y
df3 = df3.drop(columns=[col for col in df3.columns if 'run_rate' in col], errors='ignore')
df3 = df3.merge(strike_rate_df, on='player', how='left')

# Fill missing run rates with 0
df3['run_rate'] = df3['run_rate'].fillna(0)

# Optional: Clean names back to title case
df3['player'] = df3['player'].str.title()

# Display the updated df3
print(df3.head())
```

```python
# Standardize names
df3['player'] = df3['player'].str.strip().str.lower()
df['bowler'] = df['bowler'].str.strip().str.lower()

# Step 1: Total runs conceded (only runs off bat)
runs_conceded = df.groupby('bowler')['runs_of_bat'].sum()

# Step 2: Valid balls bowled (exclude wides and no-balls)
valid_balls = df[(df['wide'] == 0) & (df['noballs'] == 0)].groupby('bowler').size()

# Step 3: Economy rate = runs conceded / overs
economy_rate = (runs_conceded / valid_balls) * 6

# Step 4: Merge into df3
economy_df = economy_rate.reset_index().rename(columns={'bowler': 'player', 0: 'bowling_avg'})
economy_df.columns = ['player', 'bowling_avg']

# Drop existing column and merge new data
df3 = df3.drop(columns=[col for col in df3.columns if 'bowling_avg' in col], errors='ignore')
df3 = df3.merge(economy_df, on='player', how='left')
df3['bowling_avg'] = df3['bowling_avg'].fillna(0)
df3['player'] = df3['player'].str.title()
df3
```

```python
df3
```

```python
# Step 1: Get batting order per match and innings
batting_order = df.groupby(['match_no', 'innings'])['striker'].apply(lambda x: list(dict.fromkeys(x)))

# Step 2: Assign roles based on batting position
player_roles = {}
for order in batting_order:
    for i, player in enumerate(order):
        if i < 2:
            role = 'openers'
        elif i < 4:
            role = 'top_order'
        elif i < 8:
            role = 'middle'
        else:
            role = 'tail'
        if player not in player_roles:
            player_roles[player] = role

# Step 3: Format player names in both DataFrames for consistency
df3['player'] = df3['player'].str.strip().str.lower()

# Step 4: Update the 'category' column using .map(), only where player exists in mapping
df3['category'] = df3['player'].map(player_roles).combine_first(df3['category'])

# Step 5: Optionally re-format player names back to title case
df3['player'] = df3['player'].str.title()
```
